DATOS DE POKEMON

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URL del archivo CSV (versión RAW)
url = "https://raw.githubusercontent.com/AlvaFM/PokemonDatabase/refs/heads/main/Pokemon.csv"

# Cargar el CSV en un DataFrame
df = pd.read_csv(url)

# Mostrar las primeras filas para confirmar que se ha cargado correctamente
print(df.head())

# --------------------------- Histograma ---------------------------
# Histograma de HP
plt.figure(figsize=(10,6))
plt.hist(df['HP'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribución de HP de los Pokémon')
plt.xlabel('HP')
plt.ylabel('Frecuencia')
plt.show()

# --------------------------- Boxplot ---------------------------
# Boxplot de HP
plt.figure(figsize=(10,6))
sns.boxplot(x=df['HP'], color='skyblue')
plt.title('Boxplot de HP de los Pokémon')
plt.xlabel('HP')
plt.show()

# --------------------------- Mapa de calor (Heatmap) ---------------------------
# Correlación entre las columnas numéricas
plt.figure(figsize=(10,6))
# Select only numerical features for correlation calculation
numerical_df = df.select_dtypes(include=['number'])
correlation = numerical_df.corr()  # Calcula la correlación entre las columnas numéricas
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Mapa de Calor de Correlaciones')
plt.show()

# --------------------------- Scatterplot ---------------------------
# Scatterplot entre HP y Ataque
plt.figure(figsize=(10,6))
plt.scatter(df['HP'], df['Attack'], color='orange', edgecolor='black')
plt.title('Relación entre HP y Ataque')
plt.xlabel('HP')
plt.ylabel('Ataque')
plt.show()

# --------------------------- Gráfico de líneas ---------------------------
# Gráfico de líneas de HP a lo largo de la Generación
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Generation', y='HP', marker='o', color='green')
plt.title('Evolución de HP por Generación')
plt.xlabel('Generación')
plt.ylabel('HP')
plt.show()
```
